```python
%matplotlib inline

import pickle as pkl
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from tensorflow.examples.tutorials.mnist import input_data
import os
# Arrange GPUs to be used
os.environ["CUDA_DEVICE_ORDER"]="PCI_BUS_ID"
os.environ["CUDA_VISIBLE_DEVICES"]="0"
```

```python
# Leaky relu implementation
def leaky_relu(x, alpha=.01):
    return tf.nn.relu(x) -  tf.nn.relu(-x * alpha)
# Define generator
def generator(rnd_z, out_shape, reuse):
    global n_hidden_layers
    #rnd_z = tf.reshape(rnd_z, (-1,rnd_size_width,rnd_size_width,1))
    with tf.variable_scope('generator', reuse=reuse):
        # Generator architecture
        
        #deconv1 = tf.layers.conv2d_transpose(inputs=rnd_z, filters=n_filters*4, kernel_size=4, strides=2, padding='same', use_bias=False)
        #deconv1 = tf.layers.batch_normalization(deconv1)
        #deconv1 = leaky_relu(deconv1)#tf.nn.relu(deconv1)
        
        #deconv2 = tf.layers.conv2d_transpose(inputs=deconv1, filters=n_filters*2, kernel_size=4, strides=2, padding='same', use_bias=False)
        #deconv2 = tf.layers.batch_normalization(deconv2)
        #deconv2 = leaky_relu(deconv2)#tf.nn.relu(deconv2)
        
        #deconv3 = tf.layers.conv2d_transpose(inputs=deconv2, filters=n_filters, kernel_size=4, strides=2, padding='same', use_bias=False)
        #deconv3 = tf.layers.batch_normalization(deconv3)
        #deconv3 = leaky_relu(deconv3)#tf.nn.relu(deconv3)
        
        #logits = tf.contrib.layers.flatten(deconv2)
        layer1 = tf.layers.dense(rnd_z, n_hidden_layers, activation=leaky_relu)
        layer1 = tf.layers.batch_normalization(layer1)
        
        layer2 = tf.layers.dense(layer1, n_hidden_layers * 2, activation=leaky_relu)
        layer2 = tf.layers.batch_normalization(layer2)
        
        layer3 = tf.layers.dense(layer2, n_hidden_layers * 4, activation=leaky_relu)
        layer3 = tf.layers.batch_normalization(layer3)
                
        output = tf.layers.dense(layer3, out_shape, activation=tf.nn.tanh)
    return output
# Define discriminator (returns logits)
def discriminator(x, reuse):
    global n_hidden_layers
    #x = tf.reshape(x, (-1,input_width,input_width,1))
    with tf.variable_scope('discriminator', reuse=reuse):
        # Discriminator architecture
        #conv1 = tf.layers.conv2d(inputs=x, filters=n_filters, kernel_size=4, strides=2, padding='same', use_bias=False)
        #conv1 = tf.layers.batch_normalization(conv1)
        #conv1 = leaky_relu(conv1)
        
        #conv2 = tf.layers.conv2d(inputs=conv1, filters=n_filters*2, kernel_size=4, strides=2, padding='same', use_bias=False)
        #conv2 = tf.layers.batch_normalization(conv2)
        #conv2 = leaky_relu(conv2)
        
        #conv3 = tf.layers.conv2d(inputs=conv2, filters=n_filters*4, kernel_size=4, strides=2, padding='same', use_bias=False)
        #conv3 = tf.layers.batch_normalization(conv3)
        #conv3 = leaky_relu(conv3)
        
        #logits = tf.contrib.layers.flatten(conv1)
        #logits = tf.layers.dense(logits, 1, activation=None)
        
        layer1 = tf.layers.dense(x, n_hidden_layers * 2, activation=leaky_relu)
        layer1 = tf.layers.batch_normalization(layer1)
        
        #layer2 = tf.layers.dense(layer1, n_hidden_layers * 2, activation=leaky_relu)
        #layer2 = tf.layers.batch_normalization(layer2)
        
        layer3 = tf.layers.dense(layer1, n_hidden_layers, activation=leaky_relu)
        layer3 = tf.layers.batch_normalization(layer3)
        
        logits = tf.layers.dense(layer3, 1, activation=None)
        
        return logits
```

```python
# Parameters
n_hidden_layers = 128
input_width = 28
rnd_size_width = 10
input_size = input_width * input_width
rnd_z_size = rnd_size_width * rnd_size_width
smooth = 0.1
learning_rate = .001
batch_size = 100
n_epoch = 100
store_size = 10
```

```python
mnist = input_data.read_data_sets('../data/mnist/')
```

```python
# Graph
tf.reset_default_graph()
imag_plcholder = tf.placeholder(tf.float32, (None, input_size), name='image_placeholder')
fake_plcholder = tf.placeholder(tf.float32, (None, rnd_z_size), name='fake_placeholder')
g_out = generator(fake_plcholder, input_size, False)
d_logits_imag = discriminator(imag_plcholder, False)
d_logits_fake = discriminator(g_out, True)
# Losses
label_imag = tf.ones_like(d_logits_imag) * (1 - smooth) # Apply smoothing to help discriminator
label_fake = tf.zeros_like(d_logits_fake) # label is 1 for images, 0 for fake images
d_loss_imag = tf.nn.sigmoid_cross_entropy_with_logits(labels=label_imag, logits=d_logits_imag)
d_loss_fake = tf.nn.sigmoid_cross_entropy_with_logits(labels=label_fake, logits=d_logits_fake)
d_loss_total = tf.reduce_mean(d_loss_imag+d_loss_fake)
g_label = tf.ones_like(d_logits_fake) # Generator needs to fool discriminator
g_loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(labels=g_label, logits=d_logits_fake))
# Optimizers
g_vars = tf.get_collection(tf.GraphKeys.GLOBAL_VARIABLES, scope='generator')
d_vars = tf.get_collection(tf.GraphKeys.GLOBAL_VARIABLES, scope='discriminator')
g_opt = tf.train.AdamOptimizer(learning_rate).minimize(g_loss, var_list=g_vars)
d_opt = tf.train.AdamOptimizer(learning_rate).minimize(d_loss_total, var_list=d_vars)
```

```python
sample_arr = []
loss_arr = []
saver = tf.train.Saver(var_list=g_vars)
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    for e in range(n_epoch):
        iter_num = mnist.train.num_examples // batch_size
        for b in range(iter_num):
            # Get batch
            batch = mnist.train.next_batch(batch_size) # [batch_features, batch_labels]
            batch_images = 2 * batch[0] - 1 # Generator output is tanh, transform input similar to it.
            batch_images.reshape(-1, input_size)
            # Get random values
            batch_rnd = np.random.uniform(-1, 1, size=(batch_size, rnd_z_size))
            # Optimize
            _ = sess.run(d_opt, feed_dict={imag_plcholder:batch[0], fake_plcholder:batch_rnd})
            _ = sess.run(g_opt, feed_dict={fake_plcholder:batch_rnd})
        # Get loss for every epoch as an example of current batch
        d_loss_train = sess.run(d_loss_total, feed_dict={imag_plcholder:batch[0],fake_plcholder:batch_rnd})
        g_loss_train = g_loss.eval(feed_dict={fake_plcholder:batch_rnd})
        loss_arr.append([g_loss_train, d_loss_train])
        # Print
        print("Epoch: %4d of %d, Discriminator loss: %.4f, Generator loss: %.4f" %(e+1, n_epoch, d_loss_train, g_loss_train))
        # Get samples
        rnd_z = np.random.uniform(-1, 1, size=(store_size, rnd_z_size))
        sample_train = sess.run(generator(fake_plcholder, input_size, True), feed_dict={fake_plcholder:rnd_z})
        sample_arr.append(sample_train)
    # Save model we only need generator
    saver.save(sess, '../models/v2_GAN/gen.ckpt')
```

```python
# Plot loss
loss_np = np.asarray(loss_arr)
plt.cla()
plt.plot(loss_np.T[0,:], label='Generator')
plt.plot(loss_np.T[1,:], label='Discriminator')
plt.grid()
plt.title('GAN Training Losses')
plt.legend()
plt.show()
```

```python
# Plot samples for every 10 epoch
period = 10
n_rows, n_cols = n_epoch // period, store_size
fig, ax = plt.subplots(nrows=n_rows, ncols=n_cols, sharex=True, sharey=True, figsize=(n_cols+1,n_rows+1))
for row in range(n_rows):
    for col in range(n_cols):
        ax[row, col].imshow(sample_arr[row * period][col].reshape((28,28)), cmap='Greys_r')
```

```python
# Restore model to generate more images
saver = tf.train.Saver(var_list=g_vars)
with tf.Session() as sess:
    saver.restore(sess, '../models/v2_GAN/gen.ckpt')
    rnd_z = np.random.uniform(-1, 1, size=(store_size, rnd_z_size))
    sample = sess.run(generator(fake_plcholder, input_size, True), feed_dict={fake_plcholder:rnd_z})
    fig, ax = plt.subplots(nrows=1, ncols=store_size, sharex=True, sharey=True, figsize=(10,store_size+1))
    for col in range(store_size):
        ax[col].imshow(sample[col].reshape((28,28)), cmap='Greys_r')
```
